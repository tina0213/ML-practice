```python
from sklearn.datasets import load_iris
iris = load_iris()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
iris
```

```python
df=pd.DataFrame(iris["data"],columns=iris['feature_names'])
df["target"]=iris["target"]
df.to_csv("iris.csv",encoding="utf-8",index=False)
df
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.astype("float").corr(),annot=True,cmap="plasma")
```

```python
from sklearn.model_selection import train_test_split
#train_test_split-->(訓練90%,測試10%,訓練標籤90%,測試標籤10%)
x_train,x_test,y_train,y_test=train_test_split(df.drop(["target"],axis=1),df["target"],test_size=0.10)#(資料集,標籤,測試比例)
from sklearn.tree import DecisionTreeClassifier
clf= DecisionTreeClassifier(max_depth=3)
clf.fit(x_train,y_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz
g=export_graphviz(clf,out_file=None,feature_names=iris["feature_names"],class_names=iris["target_names"],
                  filled=True,special_characters=True)
graph=graphviz.Source(g)
graph
```

```python
pre=clf.predict(x_test)
print("預測結果:",list(pre))
print("真實結果:",list(y_test))
```

```python
from sklearn.metrics import accuracy_score
print("預測分數:",accuracy_score(y_test,pre))
```

```python
from sklearn.metrics import confusion_matrix
pd.DataFrame(confusion_matrix(y_test,pre))#混淆矩陣(X:預測值,Y:真實值)
```
